```python
%matplotlib inline
from miran import *
import librosa
import librosa.display as ld
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns

# FOR AUDIO SIGNALS WE USE THE TICKS AESTHETICS
#sns.set_style('ticks', {'axes.linewidth': 0.3,  'axes.edgecolor': 'black',})

sns.set_style('ticks', {'axes.linewidth': 0.2,  'axes.edgecolor': 'gray', 'axes.facecolor':'snow'})
sns.set_palette("Reds_d", 2); my_palette='Reds'; libcolor='Reds'

mpl.rc('font', **{'family':'serif', 'serif':['Times']})
mpl.rc('xtick', labelsize=8)
mpl.rc('ytick', labelsize=8)
mpl.rc('axes', labelsize=9)
mpl.rc('text', usetex=True)
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.major.width'] = 0.3
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 0.3


# PRINT IN GREYSCALE
#sns.set_palette("Oranges_r"); my_palette='Oranges_r'; libcolor='Oranges'
```

```python
# FUGUE1 Representations

afile1 = '/Users/angel/Dropbox/Apps/Texpad/Thesis/audio/Fugue1.wav'
sig1, sr1 = librosa.load(afile1, sr=44100)
D1 = librosa.amplitude_to_db(librosa.stft(sig1, n_fft=4096, hop_length=512), ref=np.max)
D2 = librosa.feature.chroma_cqt(sig1, sr=sr1, hop_length=512, n_chroma=12, bins_per_octave=36)

plt.figure(figsize=(5.16, 4), dpi=150)

plt.subplot(3,1,1)
ld.waveplot(sig1, x_axis='time', sr=44100)
#plt.title('(a) classical', fontsize=8)
plt.ylabel('amplitude')
plt.xticks([])
plt.xlabel('')

plt.subplot(3,1,2)
ld.specshow(D1, y_axis='log', x_axis='time', sr=44100, hop_length=512, cmap=libcolor)
#plt.title('(a) classical', fontsize=8)
plt.ylabel('hz.')
plt.xticks([])
plt.xlabel('')

plt.subplot(3,1,3)
ld.specshow(D2, y_axis='chroma', x_axis='time', sr=44100, hop_length=512, cmap=libcolor)
plt.yticks((0.5, 2.5, 4.5, 5.5, 7.5, 9.5, 11.5), ('c', 'd', 'e', 'f', 'g', 'a', 'b'))

#plot_chroma(D2, output_dir='/Users/angel/Desktop', hl=512)
plt.ylabel('chroma')
plt.xlabel('time (secs.)')
#    plt.tight_layout()
plt.savefig(os.path.join('/Users/angel/Dropbox/Apps/Texpad/Thesis/figures/', 'fugue_sig' + '.pdf'), format="pdf", dpi=1200)
```
