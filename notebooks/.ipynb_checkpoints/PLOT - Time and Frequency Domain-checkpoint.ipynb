```python
%matplotlib inline
from miran import *
import librosa
import librosa.display as ld
import matplotlib as mpl
import seaborn as sns

PRINT_QUALITY = 300

sns.set_style('ticks', {'axes.linewidth': 0.1,  'axes.edgecolor': 'black'})#, 'axes.facecolor':'snow'})
sns.set_palette("Greys_d", 1); my_palette='Greys'; libcolor='Greys'

mpl.rc('font', **{'family':'serif', 'serif':['Times']})
mpl.rc('xtick', labelsize=8)
mpl.rc('ytick', labelsize=8)
mpl.rc('axes', labelsize=9)
mpl.rc('text', usetex=True)
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.major.width'] = 0.3
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 0.3
```

```python
# TIME DOMAIN SIGNALS

afile1 = '/Users/angel/Dropbox/Apps/Texpad/Thesis/figures/sine.aiff'; sig1, sr1 = librosa.load(afile1, sr=44100, duration=0.1)
afile2 ='/Users/angel/Dropbox/Apps/Texpad/Thesis/figures/e.aiff'; sig2, sr2 = librosa.load(afile2, sr=44100, duration=0.1)
afile3 = '/Users/angel/Dropbox/Apps/Texpad/Thesis/figures/noise.aiff'; sig3, sr3 = librosa.load(afile3, sr=44100, duration=0.1)


plt.figure(figsize=(5.16, 2), dpi=150)

plt.subplot(1,3,1)
plt.ylim((-1,1))
plt.title('(a) sine', fontsize=8)
plt.ylabel('amplitude')
plt.xticks(range(0,5000,1102), [0,0.025,0.05,0.075,0.1])
plt.plot(sig1, linewidth=0.5)

plt.subplot(1,3,2)
plt.ylim((-1,1))
plt.title('(b) vowel', fontsize=8)
plt.yticks([])
plt.xlabel('time (secs.)')
plt.xticks(range(0,5000,1102), [0,0.025,0.05,0.075,0.1])
plt.plot(sig2, linewidth=0.5)

plt.subplot(1,3,3)
plt.ylim((-1,1))
plt.title('(c) noise', fontsize=8)
plt.yticks([])
plt.xticks(range(0,5000,1102), [0,0.025,0.05,0.075,0.1])
plt.plot(sig3, linewidth=0.5)

plt.tight_layout()
plt.savefig(os.path.join('/Users/angel/Dropbox/Apps/Texpad/Thesis/figures/', 'waveforms' + '.pdf'), format="pdf", dpi=PRINT_QUALITY)
```

```python
# SIGNALS SPECTRA

D1 = librosa.stft(sig1, n_fft=8192, hop_length=256)
D2 = librosa.stft(sig2, n_fft=8192, hop_length=256)
D3 = librosa.stft(sig3, n_fft=8192, hop_length=256)


plt.figure(figsize=(5.16, 2), dpi=150)

plt.subplot(1,3,1)
ld.specshow(D1, y_axis='linear', x_axis='time', sr=44100, hop_length=256, cmap=libcolor)
plt.xticks([0,0.025,0.05,0.075,0.1])
plt.ylabel('hz.')
plt.ylim((0, 1000))
plt.xlabel('')

plt.subplot(1,3,2)
ld.specshow(D2, y_axis='linear', x_axis='time', sr=44100, hop_length=256, cmap=libcolor)
plt.xticks([0,0.025,0.05,0.075,0.1])
plt.yticks([])
plt.ylabel('')
plt.ylim((0, 1000))
plt.xlabel('time (secs.)')

plt.subplot(1,3,3)
ld.specshow(D3, y_axis='linear', x_axis='time', sr=44100, hop_length=256, cmap=libcolor)
plt.yticks([])
plt.xticks([0,0.025,0.05,0.075,0.1])
plt.ylabel('')
plt.ylim((0, 1000))
plt.xlabel('')

plt.tight_layout()
plt.savefig(os.path.join('/Users/angel/Dropbox/Apps/Texpad/Thesis/figures/', 'wavespectra' + '.pdf'), format="pdf", dpi=PRINT_QUALITY)
```

```python
# FUGUE1 Representations

afile1 = '/Users/angel/Dropbox/Apps/Texpad/Thesis/audio/Fugue1.wav'
sig1, sr1 = librosa.load(afile1, sr=44100)
D1 = librosa.amplitude_to_db(librosa.stft(sig1, n_fft=4096, hop_length=512), ref=np.max)
D2 = librosa.feature.chroma_cqt(sig1, sr=sr1, hop_length=512, n_chroma=12, bins_per_octave=36)

plt.figure(figsize=(5.16, 4.5), dpi=150)

plt.subplot(3,1,1)
ld.waveplot(sig1, x_axis='time', sr=44100)
#plt.title('(a) classical', fontsize=8)
plt.ylabel('amplitude')
plt.xticks([])
plt.xlabel('')

plt.subplot(3,1,2)
ld.specshow(D1, y_axis='log', x_axis='time', sr=44100, hop_length=512, cmap=libcolor)
#plt.title('(a) classical', fontsize=8)
plt.ylabel('hz.')
plt.xticks([])
plt.xlabel('')

plt.subplot(3,1,3)
ld.specshow(D2, y_axis='chroma', x_axis='time', sr=44100, hop_length=512, cmap=libcolor)
plt.yticks((0.5, 2.5, 4.5, 5.5, 7.5, 9.5, 11.5), ('c', 'd', 'e', 'f', 'g', 'a', 'b'))

#plot_chroma(D2, output_dir='/Users/angel/Desktop', hl=512)
plt.ylabel('chroma')
plt.xlabel('time (secs.)')
plt.tight_layout()
plt.savefig(os.path.join('/Users/angel/Dropbox/Apps/Texpad/Thesis/figures/', 'fugue_sig' + '.pdf'), format="pdf", dpi=PRINT_QUALITY)
```
