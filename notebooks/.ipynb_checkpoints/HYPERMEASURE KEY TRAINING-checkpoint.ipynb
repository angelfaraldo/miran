```python
from __future__ import print_function
import madmom as mmm
from miran import *
from subprocess import call
from time import time
from sklearn import neighbors, datasets, svm
```

```python
# DECLARE PATHS
sa_path = '/Users/angel/Git/miran/scripts/sonic-annotator'
splits = folderfiles('/Users/angel/Desktop/C2/')
truth = folderfiles('/Users/angel/Insync/Datasets/beatport/keys_v2/', ext='.txt')

# INIT PLACEHOLDERS
condition         = False
harmmean          = None
rotate            = False
condition_thres   = 0
itime             = time()
idx               = 0
gkey              = []
filenames         = []
harmonicity_means = []
harmonicity_stds  = []
chroma_means      = []
chroma_stds       = []


for f in splits:
    if any(soundfile_type in f for soundfile_type in AUDIO_FILE_EXTENSIONS):
        print("Analysing '{}'".format(f), end='')
        t = time()
        sf = strip_filename(f)
        fname, fext = os.path.splitext(f)
        fdir, fname = os.path.split(fname)
        fname = os.path.join(fdir, str(idx))
        f2 = fname + fext
        os.rename(f, f2)
        
        if condition:
            # CALCULATE HARMONICITY TO DISCARD SOME SEGMENTS
            call('{0}/sonic-annotator -r -t {0}/harmonicity.n3 -w csv --csv-force "{1}"'.format(sa_path, f2), shell=True)
            harmonicity = pd.read_csv(os.path.splitext(f2)[0] + '_vamp_vamp-libxtract_spectral_inharmonicity_spectral_inharmonicity.csv', header=None)
            harmmean = harmonicity.mean()[1]
            harmstd = harmonicity.std()[1]
            os.remove(os.path.splitext(f2)[0] + '_vamp_vamp-libxtract_spectral_inharmonicity_spectral_inharmonicity.csv')

        # SET A THRESHOLD FOR DISCARDED SEGMENTS
        if harmmean is None or harmmean > condition_thres:
            
            # CHROMAGRAM ANALYSIS
            call('{0}/sonic-annotator -r -t {0}/nnls.n3 -w csv --csv-force "{1}"'.format(sa_path, f2), shell=True)
            nnls = pd.read_csv(os.path.splitext(f2)[0] + '_vamp_nnls-chroma_nnls-chroma_chroma.csv', header=None)
            mean_chroma = np.roll(nnls.drop(0,1).median(), -3)
            std_chroma  = np.roll(nnls.drop(0,1).std(), -3)
            os.remove(os.path.splitext(f2)[0] + '_vamp_nnls-chroma_nnls-chroma_chroma.csv')
            find_annotation = 0

            # APPEND GROND TRUTH LABEL
            for ff in truth:
                sff = strip_filename(ff)
                if sff == sf[:sf.rfind(' - ')]:
                    find_annotation = 1
            
                    with open(ff, 'r') as estkey:
                        key2 = None
                        key_str = estkey.read()
            
                        if '|' in key_str:
                            key_str, key2 = key_str.split('|')
                            key2 = key2.strip()                            
                            tonic_pc2 = chroma_to_pc(key2[:key2.find(' ')])
                            mode_symbol2 = key2[key2.find(' '):]

                        tonic_pc = chroma_to_pc(key_str[:key_str.find(' ')])
                        mode_symbol = key_str[key_str.find(' '):]
                        
                    if not rotate or 'X' in key_str:
                        filenames.append(os.path.split(f)[1])
                        if condition:
                            harmonicity_means.append(harmmean)
                            harmonicity_stds.append(harmstd)
                        chroma_means.append(mean_chroma)
                        gkey.append(key_str)
                                            
                    else: 
                        # TRANSPOSE TO TWELVE POSSIBLE TONICS
                        for i in range(len(mean_chroma)):
                            filenames.append(os.path.split(f)[1])
                            if condition:
                                harmonicity_means.append(harmmean)
                                harmonicity_stds.append(harmstd)
                            shift = np.roll(mean_chroma, i)
                            chroma_means.append(shift)
                            if key2:
                                transp = pc_to_chroma((tonic_pc + i) % 12) + mode_symbol + ' | ' + pc_to_chroma((tonic_pc2 + i) % 12) + mode_symbol2
                            else:
                                transp = pc_to_chroma((tonic_pc + i) % 12) + mode_symbol
                            gkey.append(transp)

            if find_annotation == 0:
                print("COULD NOT FIND ANNOTATION FOR {}".format(sf))
                
        os.rename(f2, f)    
        print(" in {} seconds.".format(time() - t))
        idx += 1

# CREATE A PD DATAFRAME WITH THE RESULTS, AND SAVE IT TO DISK
output_df = pd.DataFrame(chroma_means, index=filenames)
output_df['gkey'] = gkey

if condition:
    output_df['harmean'] = harmonicity_means
    output_df['harstd'] = harmonicity_stds

df_to_excel(output_df, '/Users/angel/Desktop/BPCONF')

print("TASK COMPLETED in {} seconds.".format(time() - itime))
```
