```python
%matplotlib inline
from miran import *
import librosa
import librosa.display as ld
import matplotlib as mpl
import matplotlib.pyplot as plt


sns.set_style('ticks', {'axes.linewidth': 0.2,  'axes.edgecolor': 'gray', 'axes.facecolor':'snow'})
sns.set_palette("Reds_d", 2); my_palette='Reds'; libcolor='Reds'
mpl.rc('font', **{'family':'serif', 'serif':['Times']})
mpl.rc('xtick', labelsize=8)
mpl.rc('ytick', labelsize=8)
mpl.rc('axes', labelsize=9)
mpl.rc('text', usetex=True)
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.major.width'] = 0.3
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 0.3

```

```python
afile1 = '/Users/angel/Dropbox/Apps/Texpad/Thesis/figures/sine.aiff'
sig1, sr1 = librosa.load(afile1, sr=44100, duration=0.1)
afile2 ='/Users/angel/Dropbox/Apps/Texpad/Thesis/figures/e.aiff'
sig2, sr2 = librosa.load(afile2, sr=44100, duration=0.1)
afile3 = '/Users/angel/Dropbox/Apps/Texpad/Thesis/figures/noise.aiff'
sig3, sr3 = librosa.load(afile3, sr=44100, duration=0.1)


plt.figure(figsize=(5.16, 2), dpi=150)

plt.subplot(1,3,1)
plt.ylim((-1,1))
plt.title('(a) sine', fontsize=8)
plt.ylabel('amplitude')
plt.xticks(range(0,5000,1102), [0,0.025,0.05,0.075,0.1])
plt.plot(sig1, linewidth=0.5)

plt.subplot(1,3,2)
plt.ylim((-1,1))
plt.title('(b) vowel', fontsize=8)
plt.yticks([])
plt.xlabel('time (secs.)')
plt.xticks(range(0,5000,1102), [0,0.025,0.05,0.075,0.1])
plt.plot(sig2, linewidth=0.5)

plt.subplot(1,3,3)
plt.ylim((-1,1))
plt.title('(c) noise', fontsize=8)
plt.yticks([])
plt.xticks(range(0,5000,1102), [0,0.025,0.05,0.075,0.1])
plt.plot(sig3, linewidth=0.5)

plt.tight_layout()
plt.savefig(os.path.join('/Users/angel/Dropbox/Apps/Texpad/Thesis/figures/', 'waveforms' + '.pdf'), format="pdf", dpi=1200)
```

```python
D1 = librosa.stft(sig1, n_fft=8192, hop_length=256)
D2 = librosa.stft(sig2, n_fft=8192, hop_length=256)
D3 = librosa.stft(sig3, n_fft=8192, hop_length=256)

plt.figure(figsize=(5.16, 2), dpi=150)

plt.subplot(1,3,1)
ld.specshow(D1, y_axis='linear', x_axis='time', sr=44100, hop_length=256, cmap=libcolor)
plt.xticks([0,0.025,0.05,0.075,0.1])
plt.ylabel('hz.')
plt.ylim((0, 1000))
plt.xlabel('')

plt.subplot(1,3,2)
ld.specshow(D2, y_axis='linear', x_axis='time', sr=44100, hop_length=256, cmap=libcolor)
plt.xticks([0,0.025,0.05,0.075,0.1])
plt.yticks([])
plt.ylabel('')
plt.ylim((0, 1000))
plt.xlabel('time (secs.)')

plt.subplot(1,3,3)
ld.specshow(D3, y_axis='linear', x_axis='time', sr=44100, hop_length=256, cmap=libcolor)
plt.yticks([])
plt.xticks([0,0.025,0.05,0.075,0.1])
plt.ylabel('')
plt.ylim((0, 1000))
plt.xlabel('')

plt.tight_layout()
plt.savefig(os.path.join('/Users/angel/Dropbox/Apps/Texpad/Thesis/figures/', 'wavespectra' + '.pdf'), format="pdf", dpi=1200)
```
